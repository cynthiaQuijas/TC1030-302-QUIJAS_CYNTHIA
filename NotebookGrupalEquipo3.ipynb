```python
# Identifica los valores atípicos utilizando un criterio específico
Q1 = df2["PESO_NETO"].quantile(0.25)
Q3 = df2["PESO_NETO"].quantile(0.75)
IQR = Q3 - Q1
filtro = (df2["PESO_NETO"] >= Q1 - 1.5 * IQR) & (df2["PESO_NETO"] <= Q3 + 1.5 * IQR)
# Selecciona solo los valores que no son atípicos y asignalos a una nueva variable
df_sin_atipicos = df2.loc[filtro]
df_sin_atipicos
```

#3.- Transformación de datos

### a) Revisa si es necesario discretizar los datos (binning) 

```python
df2['PESO_NETO'] = pd.cut(df2['PESO_NETO'], bins=np.linspace(start=df2['PESO_NETO'].min(), stop=df2['PESO_NETO'].max(), num=10))
```

```python
df2['CAP_MAXIMA'] = pd.cut(df2['CAP_MAXIMA'], bins=np.linspace(start=df2['CAP_MAXIMA'].min(), stop=df2['CAP_MAXIMA'].max(), num=5))
```

###b) Si es necesario escala y normaliza los datos.

```python
# Conversión de los valores discretizados en valores numéricos
df2['PESO_NETO'] = df2['PESO_NETO'].cat.codes
df2['CAP_MAXIMA'] = df2['CAP_MAXIMA'].cat.codes
#df['CANT_PROGRAMADA'] = df['CANT_PROGRAMADA'].apply(lambda x: (x.left + x.right)/2)
#df['PESO_NETO'] = df['PESO_NETO'].apply(lambda x: (x.left + x.right)/2)
#df['CAP_MAXIMA'] = df['CAP_MAXIMA'].apply(lambda x: (x.left + x.right)/2)
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Min-Max Scaling
scaler_minmax = MinMaxScaler()
df2[[ 'PESO_NETO', 'CAP_MAXIMA']] = scaler_minmax.fit_transform(df2[[ 'PESO_NETO', 'CAP_MAXIMA']])

# Z-Score Scaling
scaler_zscore = StandardScaler()
df2[['PESO_NETO', 'CAP_MAXIMA']] = scaler_zscore.fit_transform(df2[[ 'PESO_NETO', 'CAP_MAXIMA']])
```

###c) Si lo consideras conveniente, construye nuevos atributos que se puedan obtener a partir de los datos disponibles. (atributos derivados).

```python
#import pandas as pd
from datetime import datetime

# Convertir las columnas de fecha de presentación y salida a formato datetime
df2['F_PRESENTACION'] = pd.to_datetime(df2['F_PRESENTACION'], format='%d-%m-%Y %H:%M:%S', errors='coerce')
df2['F_PESAJESALIDA'] = pd.to_datetime(df2['F_PESAJESALIDA'], format='%d-%m-%Y %H:%M:%S', errors='coerce')

# Crear nueva columna de tiempo de espera en segundos
df2['TIEMPO_ESPERA'] = ((df2['F_PESAJESALIDA'] - df2['F_PRESENTACION']).dt.total_seconds())/3600
```
